```python
import json
import os
import pandas as pd
```

```python
data_dir = './data'
file = 'entities.json'
office_file = './belong_to.xlsx'
office_path = os.path.join(data_dir,office_file)
data_path = os.path.join(data_dir,file)
```

```python
office_df = pd.read_excel(office_path)
office_list = office_df.iloc[:,2].to_list()
office_list
```

```python
with open(data_path,'r',encoding='UTF-8') as f:
    nodes = json.load(f)
for node in nodes:
    label = node['label']
    if label == 'Department':
        name = node['name']
        if name in office_list:
            node['label'] = 'Office'
```

```python
with open('./new_entities.json','w') as f:
    json.dump(nodes,f,indent=2,ensure_ascii=False)
    print('finish')
```

```python
data_dir = './data'
file = 'new_entities.json'
recommend_eat_file = './recommand_eat.xlsx'
recommend_eat_path = os.path.join(data_dir,recommend_eat_file)
data_path = os.path.join(data_dir,file)
```

```python
recommend_eat_df = pd.read_excel(recommend_eat_path)
recommend_eat_list = recommend_eat_df.iloc[:,2].to_list()
recommend_eat_list
```

```python
with open(data_path,'r',encoding='UTF-8') as f:
    nodes = json.load(f)
for node in nodes:
    label = node['label']
    if label == 'Food':
        name = node['name']
        if name in recommend_eat_list:
            node['label'] = 'Dish'
```

```python
with open('./new_entities.json','w') as f:
    json.dump(nodes,f,indent=2,ensure_ascii=False)
    print('finish')
```

```python
relation_file = 'relations.json'
relation_path = os.path.join(data_dir,relation_file)
with open(relation_path,'r') as f:
    relations = json.load(f)
for rel in relations:
    end_entity_type=rel['end_entity_type']
    if end_entity_type == 'Department':
        rels = rel['rels']
        if rels[0]['end_entity_name'] in office_list:
            rel['end_entity_type'] = 'Office'
    if end_entity_type == 'Food':
        rels = rel['rels']
        if rels[0]['end_entity_name'] in recommend_eat_list:
            rel['end_entity_type'] = 'Dish'
```

```python
with open('./new_relations.json','w') as f:
    json.dump(relations,f,indent=2,ensure_ascii=False)
    print('finish')
```
